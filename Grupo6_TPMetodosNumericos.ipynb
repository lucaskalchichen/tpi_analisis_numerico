##### Parte 1: Derivación Numérica

**Objetivo:**  
Implementar los métodos de diferencias finitas hacia adelante, hacia atrás y centradas.

**Consigna:**  
Crear funciones en Python para cada método y aplicarlas a la función.



**Análisis:**  
Comparar los resultados obtenidos con la derivada analítica y graficar el error relativo para cada método.

---




```python
# importo librerias
import numpy as np
import matplotlib.pyplot as plt

# Defino la función

def f(x):
    return ( (3*x -1) /( x**2 + 3 ))**2

# Defino la derivada de la función
def d_primera_f(x):
    return (-18 + 50*x + 18 * x**2 -18 * x**3 )/(x**2 + 3)**3

# Metodos de diferenciacións 1


# diferencia a un pasa asi adelante
def df_hacia_adelante_1p(f,x,h):
    return (f(x+h) - f(x))/h

# diferencia a un paso hacia atras
def df_hacia_atras_1p(f,x,h):
    return (f(x) - f(x-h))/h

# diferencia a un paso central
def df_central_1p(f,x,h):
    return (f(x+h) - f(x-h))/(2*h)

# diferencia a dos pasos hacia adelante
def df_hacia_adelante_2p(f,x,h):
    return (- f(x+2*h) + 4*f(x+h) -3*f(x) )/(2*h)

# diferencia a dos pasos hacia atras

def df_hacia_atras_2p(f,x,h):
    return (3*f(x) - 4*f(x-h) + f(x-2*h) )/(2*h)

# diferencia a dos pasos central
def df_central_2p(f,x,h):
    return (-f(x+2*h) +8*f(x+h) - 8*f(x-h) + f(x+2*h))/(12*h)

# parametros a utilizar
x = 1
h = 0.1 

# calculo de la derivada de la función en x

derivada_calculada = d_primera_f(x)

# calculo las aproximaciones de la derivada en x a un paso 

dfa_adelante_p1 = df_hacia_adelante_1p(f,x,h)
dfa_atras_p1 = df_hacia_atras_1p(f,x,h)
dfa_central_p1 = df_central_1p(f,x,h)

# calculo las aproximaciones de la derivada en x a dos pasos
dfa_adelante_p2 = df_hacia_adelante_2p(f,x,h)
dfa_atras_p2 = df_hacia_atras_2p(f,x,h)
dfa_central_p2 = df_central_2p(f,x,h)

# calculo el error absuto de las aproximaciones a un paso

error_absoluto_adelante_p1 = np.abs(derivada_calculada - dfa_adelante_p1)
error_absoluto_atras_p1 = np.abs(derivada_calculada - dfa_atras_p1)
error_absoluto_central_p1 = np.abs(derivada_calculada - dfa_central_p1)

# calculo el error absoluto de las aproximaciones a dos pasos
error_absoluto_adelante_p2 = np.abs(derivada_calculada - dfa_adelante_p2)
error_absoluto_atras_p2 = np.abs(derivada_calculada - dfa_atras_p2)
error_absoluto_central_p2 = np.abs(derivada_calculada - dfa_central_p2)

# calculo el error relativo de las aproximaciones a un paso
error_relativo_adelante_p1 = (error_absoluto_adelante_p1/derivada_calculada)
error_relativo_atras_p1 = (error_absoluto_atras_p1/derivada_calculada)
error_relativo_central_p1 = (error_absoluto_central_p1/derivada_calculada)

# calculo el error relativo de las aproximaciones a dos pasos
error_relativo_adelante_p2 = (error_absoluto_adelante_p2/derivada_calculada)
error_relativo_atras_p2 = (error_absoluto_atras_p2/derivada_calculada)
error_relativo_central_p2 = (error_absoluto_central_p2/derivada_calculada)

# imprimo los resultados
print("Derivada calculada en x = ", derivada_calculada)
print("-----------------------------------------------------------------------------------------")
print("Aproximaciones a un paso")
print("-----------------------------------------------------------------------------------------")
print("Diferencia hacia adelante a un paso = ", dfa_adelante_p1)
print("Error relativo = ", error_relativo_adelante_p1)
print("              -------------                 ---------------")
print("Diferencia hacia atras a un paso = ", dfa_atras_p1)
print("Error relativo = ", error_relativo_atras_p1)
print("              -------------                 ")
print("Diferencia central a un paso = ", dfa_central_p1)
print("Error relativo = ", error_relativo_central_p1)
print("----------------------------------------------------------------------------------------")
print("Aproximaciones a dos pasos")
print("----------------------------------------------------------------------------------------")
print("Diferencia hacia adelante a dos pasos = ", dfa_adelante_p2)
print("Error relativo = ", error_relativo_adelante_p2)
print("              -------------                 ---------------")
print("Diferencia hacia atras a dos pasos = ", dfa_atras_p2)
print("Error relativo = ", error_relativo_atras_p2)
print("              -------------                 ---------------")
print("Diferencia central a dos pasos = ", dfa_central_p2)
print("Error relativo = ", error_relativo_central_p2)



h_lista = np.linspace(0.1,0.0001,10)

# grafico de la función de los errores relativos

error_adelante_p1_data = [ np.abs(d_primera_f(x) - df_hacia_adelante_1p(f,x,h))/d_primera_f(x) for h in h_lista]
error_atras_p1_data = [np.abs(d_primera_f(x) - df_hacia_atras_1p(f,x,h))/d_primera_f(x) for h in h_lista]
error_central_p1_data = [np.abs(d_primera_f(x) - df_central_1p(f,x,h))/d_primera_f(x) for h in h_lista]
error_adelante_p2_data = [np.abs(d_primera_f(x) - df_hacia_adelante_2p(f,x,h))/d_primera_f(x) for h in h_lista]
error_atras_p2_data = [np.abs(d_primera_f(x) - df_hacia_atras_2p(f,x,h))/d_primera_f(x) for h in h_lista]
error_central_p2_data = [np.abs(d_primera_f(x) - df_central_2p(f,x,h))/d_primera_f(x) for h in h_lista]

plt.plot(h_lista,error_adelante_p1_data, label = "adelante a un paso")
plt.plot(h_lista,error_atras_p1_data, label = "atras a un paso")
plt.plot(h_lista,error_central_p1_data, label = "central a un paso")
plt.plot(h_lista,error_adelante_p2_data, label = "adelante a dos pasos")
plt.plot(h_lista,error_atras_p2_data, label = "atras a dos pasos")
plt.plot(h_lista,error_central_p2_data, label = "central a dos pasos")
plt.xlabel("h")
plt.ylabel("Error relativo")
plt.legend()




```

##### Parte 2: Integración Numérica

**Objetivo:**  
Implementar la regla del trapecio (con intervalos múltiples) y las reglas de Simpson 1/3 y 3/8.

**Consigna:**  
Programar cada método y aplicarlo para calcular.



**Análisis:**

```python
# importo librerias
import numpy as np
import scipy.integrate as spi

# Defino la función a integrar

def f(x):
    return ( x**3 )/( 1 + np.sqrt(x) )

# Defino los limites de integración
a = 1
b = 2

# Calculo la integral de la función
integral = spi.quad(f,a,b)

# integral por regla del trapecio f = función a integrar, a = limite inferior, b = limite superior, n = número de particiones

def trapecio(f,a,b,h):
    
    # calculo el numero de particiones
    n = (b-a)/h
    n = int(n)
    # genero los puntos de la partición
    x = np.linspace(a,b,n+1)
    # evalúo la función en los puntos de la partición
    y = f(x)
    # multiplico por 2 los valores de la función en los puntos interiores    
    for i in range(1,n):
        y[i] = y[i]*2
        
    #sumartoria de los valores de la función en los puntos de la partición
    sumatoria = np.sum(y)
    # aproximacion de la integla por regla del trapecio
    a_integral = (n/2)*sumatoria
    
    return a_integral


# integral por regla de simpson 1/3

def simpson_con_1_3(f,a,b,h):
    
    # calculo la cantidad de particiones
    n = (b-a)/h
    n = int(n)
    
    if (n % 2 ) == 1:
        print("El número de particiones debe ser par")
        print("cambiar el ancho de la particion h")
        print("se ajustará el número de particiones")
        print("cambiando el ancho de la partición h")
        n= n + 1
        return None
    
    # genero los puntos de la partición
    x = np.linspace(a,b,n+1)
    # evalúo la función en los puntos de la partición
    y = f(x)
    
    # calulos n+1 puntos de la partición
    y4 = y[1:n:2]
    
    # calculos n puntos de la partición
    y2 = y[2:n-1:2]
   
    # sumatoria de los valores de la función en los puntos de la partición
    
    sumatoria_2 = np.sum(y2)*2
    sumatoria_4 = np.sum(y4)*4
    
    # aproximación de la integral por regla de simpson 1/3
    integral = (h/6)*(y[0] + sumatoria_2 + sumatoria_4 + y[-1])
    
    return integral

# integral por regla de simpson 3/8

def simpson_con_3_8(f,a,b,h):
    
    # calculo la cantidad de particiones
    n = (b-a)/h
    n = int(n)
    
    if n % 2 == 0:
        print("El número de particiones debe ser impar")
        print("cambiar el ancho de la particion h")
        print("se ajustará el número de particiones")
        print("cambiando el ancho de la partición h")
        n = n + 1
        
    
    
    # genero los puntos de la partición
    x = np.linspace(a,b,n+1)
    # evalúo la función en los puntos de la partición
    y = f(x)
    
    # calulos puntos de la partición interiores
    y3 = y[2:n-1]
    
    # sumatoria de los valores de la función en los puntos de la partición
    
    sumatoria_3 = np.sum(y3)*3
    
    
    # aproximación de la integral por regla de simpson 1/3
    integral = ( h*8 / 3 )*(y[0] + sumatoria_3 + y[-1])
    
    return integral
    
# parametros a utilizar
h = 0.25

# calculo de la integral por regla del trapecio
integral_trapecio = trapecio(f,a,b,h)

# calculo de la integral por regla de simpson 1/3
integral_simpson_1_3 = simpson_con_1_3(f,a,b,h)

# calculo de la integral por regla de simpson 3/8
integral_simpson_3_8 = simpson_con_3_8(f,a,b,h)

# imprimo los resultados
print("Integral calculada = ", integral)
print("-----------------------------------------------------------------------------------------")
print("Aproximaciones")
print("\n")

print("Regla del trapecio = ", integral_trapecio)
print("Error absoluto = ", np.abs(integral[0] - integral_trapecio))
print("\n")
print("Regla de simpson 1/3 = ", integral_simpson_1_3)
print("Error absoluto = ", np.abs(integral[0] - integral_simpson_1_3))
print("\n")
print("Regla de simpson 3/8 = ", integral_simpson_3_8)
print("Error absoluto = ", np.abs(integral[0] - integral_simpson_3_8))
```
